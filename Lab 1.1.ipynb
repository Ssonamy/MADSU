```python
df['sex'].value_counts()
```

```python
df['sex'].value_counts(dropna=False)
```

Выведем самое распространённое значение в столбце "sex":

```python
top_sex = df['sex'].value_counts().index[0]
top_sex
```

```python
df['sex'] = df['sex'].fillna(top_sex)
df['sex'].value_counts()
```
